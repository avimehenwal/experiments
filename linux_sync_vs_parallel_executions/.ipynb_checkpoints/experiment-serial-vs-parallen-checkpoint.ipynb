```python
def human_2_miliseconds (timestamp):
    canonical_timestamp = timestamp.strip().split('m')
    minutes = int(canonical_timestamp[0])
    mseconds = int(float(canonical_timestamp[-1].split('s')[0]) * 1000)
    total_ms = (minutes * 60000 ) + mseconds
    return total_ms
```

```python
x_axis_range = [ i for i in range(1, 11) ]
x_axis_range
```

```python
data_serial_plot = [ human_2_miliseconds(sample['real']) for sample in serial_execution_samples ]
data_serial_plot
```

```python
plt.plot(x_axis_range, data_serial_plot)
plt.xlabel('Number of Samples')
plt.ylabel('Real execution time in ms')
plt.title("Serialized execution")
plt.show()
```

```python
data_parallel_plot = [ human_2_miliseconds(sample['real']) for sample in parallel_execution_samples ]
data_parallel_plot
```

```python
plt.plot(x_axis_range, data_parallel_plot)
plt.xlabel('Number of Samples')
plt.ylabel('Real execution time in ms')
plt.title("Parallel execution")
plt.show()
```

```python
plt.plot(x_axis_range, data_serial_plot, color='b', label='serial')
plt.plot(x_axis_range, data_parallel_plot, color='r', label='parallel')
plt.xlabel('Number of Samples')
plt.ylabel('Real execution time in ms')
plt.title("serial vs parallen execution")
plt.show()
```

```python
data_analysis = pd.DataFrame({
    'synchronous_real': [ human_2_miliseconds(sample['real']) for sample in serial_execution_samples ],
    'synchronous_user': [ human_2_miliseconds(sample['user']) for sample in serial_execution_samples ],
    'synchronous_sys': [ human_2_miliseconds(sample['sys']) for sample in serial_execution_samples ],
 
    'parallel_real': [ human_2_miliseconds(sample['real']) for sample in parallel_execution_samples ],
    'parallel_user': [ human_2_miliseconds(sample['user']) for sample in parallel_execution_samples ],
    'parallel_sys': [ human_2_miliseconds(sample['sys']) for sample in parallel_execution_samples ],
 })
data_analysis
```

```python
data_analysis.plot()
plt.savefig('analysis_results.png', dpi=1000)
plt.show()
```

```python
synchronous_mean = statistics.mean([ human_2_miliseconds(sample['real']) for sample in serial_execution_samples ])
parallel_mean = statistics.mean([ human_2_miliseconds(sample['real']) for sample in parallel_execution_samples ])

times_faster = synchronous_mean / parallel_mean
abs_times_faster = synchronous_mean // parallel_mean

print(synchronous_mean, parallel_mean, times_faster, abs_times_faster)

# (parallel_mean / synchronous_mean ) * 100

times_faster = [ parallel_mean, synchronous_mean,]

plt.plot([1, 2], times_faster, color='b', label='boost')
plt.xlabel('Number of Samples')
plt.ylabel('Real execution time in ms')
plt.title("serial vs parallen execution")
plt.show()
# times_faster.plot.bar(stacked=True)
# plt.savefig('gain_analysis_results.png', dpi=1000)
# plt.show()
```
